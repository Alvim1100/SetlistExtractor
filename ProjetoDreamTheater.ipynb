```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
```

- Fazer o loop pra trocar de página e trocar o url
- Botar tudo num df pandas e transformar em csv

```python
url = "https://www.setlist.fm/setlists/dream-theater-bd6a102.html?tour=13dd2169"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/146.0.0.0 Safari/537.36"}

page = requests.get(url, headers=headers)

base_soup = BeautifulSoup(page.content, "html.parser")
```

```python
def change_soup(shows, i):
    soup = BeautifulSoup(requests.get(urljoin(url, shows[i]), headers=headers).content, "html.parser")
    return soup
```

```python
def get_shows_per_page(base_soup):
    shows = []
    url_base = "https://www.setlist.fm" 
    
    for a_tag in base_soup.find_all("a", title=True, href=True):
        
        #verifica se o título é de um show (começa com "View this")
        if "View this" in a_tag['title'] and "upcoming" not in a_tag['title'].lower():
            
            incomplete_url = a_tag['href']
            full_url = urljoin(url_base, incomplete_url)

            
            if full_url not in shows:
                shows.append(full_url)
    return shows
shows = get_shows_per_page(base_soup)
for show in shows:
    print(show)
```

```python
def get_songs(shows, i):
    soup = change_soup(shows, i)
    songs = []
    for i in soup.find_all("a", class_ = "songLabel"):
        songs.append(i.text)
    return songs

print(get_songs(get_shows_per_page(base_soup), 0))
```

```python
def get_venue(shows, i):
    soup = change_soup(shows, i)
    venue = soup.find("a", href=lambda href: href and "/venue/" in href).text.strip()
    return venue

print(get_venue(get_shows_per_page(base_soup), 0))
```

```python
def get_date(shows, i):
    soup = change_soup(shows, i)
    month = soup.find("span", class_ = "month").text.strip()
    day = soup.find("span", class_ = "day").text.strip()
    year = soup.find("span", class_ = "year").text.strip()
    date = f"{month} {day}, {year}"
    return date 

print(get_date(get_shows_per_page(base_soup), 0))
```

```python
def get_band(shows, i):
    soup = change_soup(shows, i)
    band = soup.find("a", href=lambda href: href and "/setlists/" in href).text.strip()
    return band

print(get_band(get_shows_per_page(base_soup), 0))
```

```python
print(get_band(get_shows_per_page(base_soup), 0))
print("Show: ", get_venue(get_shows_per_page(base_soup), 0), " - ", get_date(get_shows_per_page(base_soup), 0))
print("Setlist: ", get_songs(get_shows_per_page(base_soup), 0))
```
